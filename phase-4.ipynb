## Storing image in Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("numpy-logo.png").convert("RGB")
logo_matrix = np.asarray(img)

np.save("numpy-logo.npy", logo_matrix)
```

```python
# Print logo
try:
    logo = np.load('numpy-logo.npy')

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)

    # Converting img into dark mode
    ## Method: 1

    dark_logo = 1-logo

    plt.subplot(1, 2, 2)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.grid(False)

except FileNotFoundError:
    print("numpy image not found")
```
