## Numpy array and basics
Learn about numpy

```python
import numpy as np
```

## Creating array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D array: ", arr_1d)

arr_2d = np.array([[1,2,3], [4,5,6]])
print("2D array: ", arr_2d)
```

### Numpy array vs List

```python
py_list = [1, 2, 3]
print("List multiplication", py_list * 2)

np_array = np.array([1, 2, 3]) #element wise multiplication
print("Array multiplication", np_array * 2)
```

## Numpy array vs List operation time

```python
import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time ", time.time() - start)

start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy operation time ", time.time() - start)
```

## Creating array from scratch

```python
#Creating zeros array
zeros = np.zeros((3,4))
print("Zeros array: \n", zeros)


# Creating ones array
ones = np.ones((3, 4))
print("Ones array: \n", ones)


#Creating constant number array
full = np.full((3,4), 5)
print("Constant array: \n", full)


#Creating random array (0 to 1)
random = np.random.random((3, 4))
print("Random array: \n", random)


#creating sequence array
seq = np.arange(0, 11, 2)
print("Sequence array: \n", seq)
```

## Vector, Matrix, and Tensor

```python
#Creating vector
vector = np.array([1, 2, 3, 4])
print("Vector: ", vector)


#Creating matrix
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Matrix: ", matrix)

#Creating tensor
tensor = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])
print("Tensor: ", tensor)
```

## Array Properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Shape: ", arr.shape)
print("Dimension: ", arr.ndim)
print("Size: ", arr.size)
print("Data Type: ", arr.dtype)
```

## Array Reshaping

```python
#Reshape an array
arr = np.arange(12)
print("Original array: ", arr)

reshaped = arr.reshape((3, 4))
print("\n Reshaped array: ", reshaped)

#Flattened array
flattened = reshaped.flatten()
print("\n Flattern array: ", flattened)

#Ravel array (returns view, instead of copy)
ravel = reshaped.ravel()
print("\n Ravel array: ", ravel)

#Tranpose array
transpose = reshaped.T
print("\n Transpose array: ", transpose)
```
