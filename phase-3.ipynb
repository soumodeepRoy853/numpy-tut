```python
import numpy as np
import matplotlib.pyplot as plt
```

## Advance operation with Business examples

```python
# Data structure: [restuarant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # restuarant A
    [2, 120000, 140000, 160000, 190000], # restuarant B
    [3, 200000, 230000, 260000, 300000], # restuarant C
    [4, 180000, 210000, 240000, 270000], # restuarant D
    [5, 160000, 185000, 205000, 230000]  # restuarant E
])
print("========== Swiggy Sales Analysis ==========")
print("Sales data shape: ", sales_data.shape)
print("\n First 3 restuarant data:\n ", sales_data[0:3])
```

```python
# Total sales per year
total_sales = np.sum(sales_data[:, 1:], axis=0) # axis=0: column-wise
print("Total sales Column wise: ", total_sales)


# Minimum sales per restuarant (Row wise)
min_sales = np.min(sales_data[:,1:], axis=1) # axis=1: row-wise
print("Minimum sales per restuarant: ", min_sales)


# Maximum sales per year
max_sales = np.max(sales_data[:,1:], axis=0) # axis=0: column-wise
print("Maximum sales per year: ", max_sales)


# Avg sales per restuarant
avg_sales = np.mean(sales_data[:,1:], axis=1)
print("Average sales per restuarant: ", avg_sales)


# Cumulative sales
cum_sales = np.cumsum(sales_data[:,1:], axis=1)
print("Cumulative sum: \n", cum_sales)

# Print figure using matplotlib
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Average cumulative sales accorss all restuarant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector Addition, Multiplication, Dot Product, Angle

```python
vec1 = np.array([1, 2, 3, 4, 5])
vec2 = np.array([6, 7, 8, 9, 10])

# Addition vector
print("Vector addition: ", vec1 + vec2)

# Multiplication vector
print("Multiplication vector: ", vec1 * vec2)

#Dot Product vector
print("Dot product: ", np.dot(vec1, vec2))

#Angle of vector
angle = np.arccos(np.dot(vec1, vec2) / (np.linalg.norm(vec1) * (np.linalg.norm(vec2))))
print("Angle vector: ", angle)
```
